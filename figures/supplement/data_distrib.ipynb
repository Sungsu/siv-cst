```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
colors = ['w']+['g' for g in range(19)]
counts = [i for i in [1, 25, 25, 7, 25, 25, 25, 6, 6, 11, 22, 5, 1, 5, 17, 7, 2, 6, 16, 3, 1, 25, 24, 1, 3, 32, 8, 25, 25, 25, 2, 25, 4] 
          if i >= 5]
sns.distplot(counts, color='g', bins=20, kde=False)
plt.xlim(0, 32)
sns.set_style('white')
plt.xlabel('Number of Sequences per Host', size=20)
plt.ylabel('Number of Hosts', size=20)
plt.xticks(size=14)
plt.yticks(size=14)
plt.tight_layout()

sns.set_style('white')
```

#### Figure S3 Distribution of the number of sequences per host used in the main analysis
All available high-quality lentivirus sequences were randomly subsampled up to 25 sequences per host. In order to represent the full breadth of known genetic diversity of lentiviruses in each host population, a small number of additional sequences were permitted for the few hosts that are infected by multiple viral lineages.
